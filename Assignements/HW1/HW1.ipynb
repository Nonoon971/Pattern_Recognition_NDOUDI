1. An identity matrix is a square matrix where the value of the diagonal coefficients is `'1'` and the others are `'0'`  
  
$$
I_{n} =
 \begin{pmatrix}
  1 & 0 & \cdots & 0 \\
  0 & 1 & \cdots & 0 \\
  \vdots  & \vdots  & \ddots & \vdots  \\
  0 & 0 & \cdots & 1
 \end{pmatrix}
$$

 Thus, the multiplication of matrix A with an identity matrix return the matrix A
 $$
 A\times I_{n} = A
 $$

```python
#Create an identity matrix that as the same size as img_average_channel
identity_matrix = np.identity(img_average_channel.shape[0])

import random
#We take a random row between [0,99] and go through all the columns of this row
row = random.randint(0, img_average_channel.shape[0] - 1)
random_row = identity_matrix[row, :]

# Print a random row out of the identy matrix
print("Random row of the identity matrix:")
print(random_row)
print(random_row[row])
```

We can see that there is only one coefficient with the value `'1'` and the others are `'0'`.  
What we can also notice is that the coefficient of the umpteenth row took randomly with the same umpteenth column is the one with value `'1'` as the definition of an identity matrix explains.

```python
#multiply T and img_average_channel
transformed_image = T @ img_average_channel
print("Shape of the multiplication between the matrix T & img_avg_channel: " + str(transformed_image.shape))

plt.imshow(transformed_image)
plt.title('Transformed Image')
plt.show()

#Test with the data found
print(transformed_image[0] == img_average_channel[47])
```

The pattern we see is the img_average_channel image but not exactly the same.

the multiplication will result to have the matrix A but with the rows in the same shuffle order as T.

So the order of `transformed_image` columns will correspond to the order of `rows` because the `'1'` value will no longer be on the diagonal but specified by the indices of `rows`. So, as explained above:

## Part 7

1. We now want to retrieve the original image. Is matrix `T` invertible? How would you show that it is invertible?
2. Compute the inverse and store the matrix in `Tinv`.
3. Perform the inverse of the scrambling operation and plot the result to show that you retrieved the original image.

```python
#Verify if T is invertible with the determinant
print("Calculation of the determinant of T = " + str(np.linalg.det((T))))
if(np.linalg.det((T)) != 0):
    print("T is invertible")
else:
    print("T is not invertible")
```

$det(T)\neq 0$ so T is invertible

```python
#Compute the inverse and store the matri
Tinv = np.linalg.inv(T)
```

To retrieve the original image, we just have to multiply the inverse matrix T with `transformed_imaged`  
In fact, $transformedImage = T\times imgAvgChannel$ and so $$Tinv\times transformedImage = T^{-1}\times T\times imgAvgChannel$$
And we know that $T^{-1}\times T = I_{n}$  

Thus, $I_{n}\times imgAvgChannel = imgAvgChannel$


```python
#Perform the inverse of the scrambling operation and plot the result to show that you retrieved the original image.
retrieved_image = Tinv @ transformed_image

plt.imshow(retrieved_image)
plt.title('Retrieved the original image')
plt.show()
```
